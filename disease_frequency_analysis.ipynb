```python
df=df.sort_values(by="Cases",ascending=False)
df
```

```python
disease_summary = df.groupby("Disease")["Cases"].sum().sort_values(ascending=False)

print(disease_summary)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
disease_summary.plot(kind="bar", color="yellow")
plt.title("Most Common Diseases Reported")
plt.xlabel("Disease")
plt.ylabel("Total Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
